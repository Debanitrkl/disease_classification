```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import MultiLabelBinarizer
```

```python
cleaned_data.drop(columns=['HEART_RATE','HEAD_CIRCUMFERENCE','UPPER_ARM_CIRCUMFERENCE'],inplace = True)
```

```python
vitals =['SYSTOLIC_BP','DIASTOLIC_BP','PULSE','RESPIRATION_RATE','BODY_TEMPERATURE','BODY_WEIGHT','HEIGHT','SPO2']
```

```python
cleaned_data[vitals] = cleaned_data[vitals].replace(0, np.nan)
#cleaned_data[['SYSTOLIC_BP','DIASTOLIC_BP']].replace(0, np.nan, inplace = True)
```

```python
cleaned_data[(cleaned_data.SYSTOLIC_BP>230) | (cleaned_data.SYSTOLIC_BP<80)].to_csv('abnormal_systolic.csv',index=False)
cleaned_data[(cleaned_data.DIASTOLIC_BP>150) | (cleaned_data.DIASTOLIC_BP<50)].to_csv('abnormal_diastolic.csv',index=False)
cleaned_data[(cleaned_data.BODY_TEMPERATURE<90) | (cleaned_data.BODY_TEMPERATURE>110)].to_csv('abnormal_body_temp.csv',index=False)
cleaned_data[(cleaned_data.SPO2>100) | (cleaned_data.SPO2<80)].to_csv('abnormal_spo2.csv',index=False)
cleaned_data[cleaned_data.AGE>100].to_csv('abnormal_age.csv',index=False)
cleaned_data[(cleaned_data.PULSE<50) | (cleaned_data.PULSE>120)].to_csv('abnormal_pulse.csv',index=False)
cleaned_data[(cleaned_data.RESPIRATION_RATE>30) | (cleaned_data.RESPIRATION_RATE<13)].to_csv('abnormal_resp_rate.csv',index=False)
cleaned_data[cleaned_data.BODY_WEIGHT>200].to_csv('abnormal_weight.csv',index=False)
cleaned_data[(cleaned_data.HEIGHT>200) | (cleaned_data.HEIGHT<45)].to_csv('abnormal_height.csv',index=False)
```

```python
def drop_wrong_vitals(df):

    df.drop(df[(df.SYSTOLIC_BP>230) | (df.SYSTOLIC_BP<80)].index, inplace = True)
    df.drop(df[(df.DIASTOLIC_BP>150) | (df.DIASTOLIC_BP<50)].index, inplace=True)
    df.drop(df[(df.BODY_TEMPERATURE<90) | (df.BODY_TEMPERATURE>110)].index, inplace=True)
    df.drop(df[(df.SPO2>100) | (df.SPO2<80)].index,inplace=True)
    df.drop(df[df.AGE>100].index, inplace=True)
    df.drop(df[(df.RESPIRATION_RATE>30) | (df.RESPIRATION_RATE<13)].index,inplace=True)
    df.drop(df[df.BODY_WEIGHT>200].index,inplace=True)
    df.drop(df[(df.HEIGHT>200) | (df.HEIGHT<45)].index,inplace=True)
    
    return df
```

```python
cleaned_data = drop_wrong_vitals(cleaned_data)
```
